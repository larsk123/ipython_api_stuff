```python
#this is start of joining data tables (inner, outer, full, etc.)
#see good youtube video on joins: https://www.youtube.com/watch?v=KTvYHEntvn8
df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],'value': np.random.randn(4)})
```

```python
df2 = pd.DataFrame({'key': ['B', 'D', 'D', 'E'],'value': np.random.randn(4)})
```

```python
# merge performs an INNER JOIN by default
pd.merge(df1, df2, on='key')
```

```python
# left outer join - show all records from df1
pd.merge(df1, df2, on='key', how='left')
```

```python
# right join - show all records from df2
pd.merge(df1, df2, on='key', how='right')
```

```python
# full join - show all records from both frames
pd.merge(df1, df2, on='key', how='outer')
```

```python
#Unions - check out site: https://stackoverflow.com/questions/905379/what-is-the-difference-between-join-and-union
#UNION ALL can be performed using concat(). Example:
df1 = pd.DataFrame({'city': ['Chicago', 'San Francisco', 'New York City'],'rank': range(1, 4)})
df2 = pd.DataFrame({'city': ['Chicago', 'Boston', 'Los Angeles'],'rank': [1, 4, 5]})
```

```python
pd.concat([df1, df2])
```

```python
#In pandas, you can use concat() in conjunction with drop_duplicates().
pd.concat([df1, df2]).drop_duplicates()
```

```python
#Pandas equivalents for some SQL analytic and aggregate functions
#Top N rows with offset
#back to the old table with tips, etc.
tips.nlargest(10+5, columns='tip').tail(10)
```

```python
#Top N rows per group
(tips.assign(rn=tips.sort_values(['total_bill'], 
             ascending=False).groupby(['day'])
             .cumcount() + 1).query('rn < 3').sort_values(['day','rn']))
```

```python
#the same using rank(method=’first’) function
#do play around with this to understand better. Rank by other stuff, etc.
(tips.assign(rnk=tips.groupby(['day'])['total_bill']
             .rank(method='first', ascending=False))
             .query('rnk < 3')
             .sort_values(['day','rnk']))
```

```python
#Let’s find tips with (rank < 3) per gender group for (tips < 2). Notice that when using rank(method='min') function rnk_min remains the same for the same tip (as Oracle’s RANK() function)
(tips[tips['tip'] < 2].assign(rnk_min=tips.groupby(['sex'])['tip']
                    .rank(method='min')).query('rnk_min < 3')
                    .sort_values(['sex','rnk_min']))
```

```python
#update. Here updating the tip
tips.loc[tips['tip'] < 2, 'tip'] *= 2
```

```python
#delete
#In pandas we select the rows that should remain, instead of deleting them
tips = tips.loc[tips['tip'] <= 9]
```
